## Unexpected Missing Values

```python
df['OWN_OCCUPIED']
```

```python
df['OWN_OCCUPIED'].isnull()
```

```python
cnt=0
for row in df['OWN_OCCUPIED']:
    try:
        int(row)
        df.loc[cnt, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    cnt+=1
df.head(9)
```

## Time Series

### Basic Time Series Manipulation

```python
from datetime import datetime
date_rng = pd.date_range(start='1/01/2020', end='1/08/2020', freq='H')
```

```python
# memiliki timestamps dengan frekuensi per jam
date_rng
```

```python
df = pd.DataFrame(date_rng, columns=['date'])
df['data'] = np.random.randint(0,100,size=(len(date_rng)))
df.head()
```

```python
# konversi indeks data frame menjadi datetime index
df['datetime'] = pd.to_datetime(df['date'])
df = df.set_index('datetime')
df.drop(['date'], axis=1, inplace=True)
df.head()
```

```python
# ubah date_rng kita menjadi list of string dan kemudian ubah string tersebut menjadi stempel waktu
string_date_rng = [str(x) for x in date_rng]
string_date_rng
```

```python
# mengonversi string menjadi timestamps 
timestamp_date_rng = pd.to_datetime(string_date_rng, infer_datetime_format=True)
timestamp_date_rng
```

```python
# buat daftar tanggal yang berubah-ubah menjadi string dan mengubahnya menjadi timestamps
string_date_rng_2 = ['June-01-2020', 'June-02-2020', 'June-03-2020']
timestamp_date_rng_2 = [datetime.strptime(x,'%B-%d-%Y') for x in string_date_rng_2]
timestamp_date_rng_2
```

```python
# memasukkan ke dataframe
df2 = pd.DataFrame(timestamp_date_rng_2, columns=['date'])
df2
```

```python
# lihat datanya dengan parsing pada timestamp index
df[df.index.day == 2]
```

```python
# memanggil tanggal yang ingin kita lihat melalui index dari data frame
df['2020-01-03']
```

```python
# memilih data di antara tanggal tertentu
df['2020-01-04':'2020-01-06']
```

```python
# menghitung rata-rata harian dari data
df.resample('D').mean()
```

```python
df['rolling_sum'] = df.rolling(3).sum()
df.head()
```

```python
# dengan kolom baru yang mengambil rolling sum dan backfills data
df['rolling_sum_backfilled'] = df['rolling_sum'].fillna(method='backfill')
df.head()
```

```python
opsd_daily = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv')
opsd_daily.shape
```

```python
#set sebagai indeks DataFrame.
opsd_daily = opsd_daily.set_index('Date')
opsd_daily.head(3)
```

```python
opsd_daily.index
```

```python
# menggabungkan langkah-langkah di atas menjadi satu baris, menggunakan parameter index_col dan parse_dates dari fungsi read_csv(). 
opsd_daily = pd.read_csv('https://raw.githubusercontent.com/ardhiraka/PFDS_sources/master/opsd_germany_daily.csv', index_col=0, parse_dates=True)
opsd_daily.head()
```

```python
# menambahkan kolom
opsd_daily['Year'] = opsd_daily.index.year
opsd_daily['Month'] = opsd_daily.index.month
opsd_daily['Weekday'] = opsd_daily.index.weekday
opsd_daily.head()
```

```python
# memilih data untuk satu hari menggunakan string seperti '2017-08-10'.

opsd_daily.loc['2017-08-10']
```

```python
# rentang waktu 
opsd_daily.loc['2014-01-20':'2014-01-22']
```

```python
opsd_daily.loc['2012-02']
```

```python
# buat line plot dari time series konsumsi listrik harian di Jerman, menggunakan method plot() DataFrame.

opsd_daily['Consumption'].plot(linewidth=0.5);
```

```python
# plot data sebagai titik-titik, dan lihat juga time series dari Matahari dan Angin.
# consumption
opsd_daily['Consumption'].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9));
```

```python
# solar
opsd_daily['Solar'].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9));
```

```python
# wind
opsd_daily['Wind'].plot(marker='.', alpha=0.5, linestyle='None', figsize=(11, 9));
```

```python
# plot deret waktu dalam satu tahun untuk menyelidiki lebih lanjut.

ax = opsd_daily.loc['2017', 'Consumption'].plot(figsize=(11, 9))
ax.set_ylabel('Daily Consumption (GWh)');
```

```python
# perbesar lebih jauh dan lihat bulan Januari dan Februari saja.

ax = opsd_daily.loc['2017-01':'2017-02', 'Consumption'].plot(marker='o', linestyle='-', figsize=(11, 9))
ax.set_ylabel('Daily Consumption (GWh)');
```

```python
# mengelompokkan data berdasarkan bulan, untuk memvisualisasikan yearly seasonality.

opsd_daily.boxplot(column=['Consumption'], by='Month')
```

```python
opsd_daily.boxplot(column=['Solar'], by='Month')
```

```python
opsd_daily.boxplot(column=['Wind'], by='Month')
```

```python
# kelompokkan rangkaian waktu konsumsi listrik berdasarkan hari dalam seminggu, untuk menjelajahi weekly seasonality.

opsd_daily.boxplot(column=['Consumption'], by='Weekday')
```

```python
# resample data menjadi weekly mean time series.

data_columns = ['Consumption', 'Wind', 'Solar', 'Wind+Solar']
opsd_weekly_mean = opsd_daily[data_columns].resample('W').mean()
opsd_weekly_mean.head(3)
```

```python
# mengonfirmasi ini dengan membandingkan jumlah baris dari dua DataFrame.

print(opsd_daily.shape[0])
print(opsd_weekly_mean.shape[0])
```

```python
# gambarkan rangkaian waktu Solar harian dan mingguan dalam satu periode enam bulan untuk membandingkannya.

start, end = '2017-01', '2017-06'

opsd_daily.loc[start:end, 'Solar'].plot(marker='.', linestyle='-', linewidth=0.5, figsize=(11, 9))
opsd_weekly_mean.loc[start:end, 'Solar'].plot(marker='o', markersize=8, linestyle='-', figsize=(11, 9))
```
